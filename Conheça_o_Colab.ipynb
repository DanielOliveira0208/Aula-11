```
from deep_translator import GoogleTranslator

# Texto para tradução
texto = "Hello, world! How are you today?"

# Traduzindo para diferentes idiomas
traducao_pt = GoogleTranslator(source='en', target='pt').translate(texto)  # Português
traducao_de = GoogleTranslator(source='en', target='de').translate(texto)  # Alemão
traducao_ru = GoogleTranslator(source='en', target='ru').translate(texto)  # Russo
traducao_it = GoogleTranslator(source='en', target='it').translate(texto)  # Italiano
traducao_ar = GoogleTranslator(source='en', target='ar').translate(texto)  # Árabe

# Exibindo as traduções
print("Texto original:", texto)
print("Tradução para Português:", traducao_pt)
print("Tradução para Alemão:", traducao_de)
print("Tradução para Russo:", traducao_ru)
print("Tradução para Italiano:", traducao_it)
print("Tradução para Árabe:", traducao_ar)
```
